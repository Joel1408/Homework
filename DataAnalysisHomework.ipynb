```python
df2=df1.loc[~df1.duplicated(subset=['Coaster_Name','Location','Opening_Date_Clean'])]
df2.shape
```

```python
#Checking the frequency for each value in the Year_introduced column
df2['Year_Introduced'].value_counts()
```

```python
#Creating a distribution plot for the column Year_Introduced
ax = df2['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Years When Rollercoasters were Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```
